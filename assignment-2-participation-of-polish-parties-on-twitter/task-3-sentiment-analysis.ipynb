```python
from polyglot.text import Text
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
parties = ["pisorgpl", "Porozumienie__", "SolidarnaPL", "platforma_org", "Nowoczesna", "Zieloni", "inicjatywaPL", 
            "KONFEDERACJA_", "RuchNarodowy", "Partia_KORWiN", "nowePSL","__Lewica", "partiarazem","kukiz15"]
```

```python
def get_text_sentiment(text):
    text = Text(text)
    text.language = 'pl'
    return text.polarity
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for party in parties:
    data = pd.read_csv(party + ".csv")
    data = data[(data.date >="2020-07-01") & (data.date <="2020-07-31")]
    sentiment_df = [] 
    for index, row in data.iterrows():
        sentiment_df.append(get_text_sentiment(row['tweet']))
    data['sentiment'] = sentiment_df
    grp = data.groupby(['date'])['sentiment'].mean()
    ax = grp.plot(ax=ax, kind='line', label=party,
                      style='-', title='Sentiment in time')
    plt.xlabel('Time')
    plt.ylabel('Sentiment')
    plt.legend()
```
